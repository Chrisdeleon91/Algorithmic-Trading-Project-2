```python
# Define the API endpoint for Fear & Greed Index with a limit of 2500 data points
url = "https://api.alternative.me/fng/?limit=1000"

# Make API request for Fear & Greed Index
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    
    fear_greed_data = pd.DataFrame(data['data'])
    
   
    fear_greed_data['timestamp'] = pd.to_datetime(fear_greed_data['timestamp'], unit='s')

    
    plt.figure(figsize=(10, 6))
    plt.plot(fear_greed_data['timestamp'], fear_greed_data['value'], marker='o', linestyle='-', color='#4287f5', markersize=8, label='Fear & Greed Index')
    
    # Style the plot
    plt.title('Fear & Greed Index Over Time', fontsize=16, fontweight='bold')
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Fear & Greed Index', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    # Beautify the x-axis date labels
    plt.xticks(rotation=45, ha='right')
    
    # Show a horizontal line at the midpoint for reference
    plt.axhline(y=50, color='gray', linestyle='--', linewidth=1, alpha=0.7)
    
    plt.tight_layout()
    plt.show()

else:
    print(f"Error: {response.status_code}")
```

```python
import requests
import datetime

# Replace 'bitcoin' with the cryptocurrency of your choice
crypto_symbol = 'bitcoin'

# Specify the date range in the format 'dd-mm-yyyy'
start_date = '17-01-2021'
end_date = '17-01-2024'

# CoinGecko API endpoint for historical prices
url = f'https://api.coingecko.com/api/v3/coins/{crypto_symbol}/market_chart'
params = {
    'vs_currency': 'usd',
    'from': int(datetime.datetime.strptime(start_date, "%d-%m-%Y").timestamp()),
    'to': int(datetime.datetime.strptime(end_date, "%d-%m-%Y").timestamp()),
    'days': (datetime.datetime.strptime(end_date, "%d-%m-%Y") - datetime.datetime.strptime(start_date, "%d-%m-%Y")).days + 1,  # Calculate the number of days
    'interval': 'daily',  # You can adjust the interval as needed
}

response = requests.get(url, params=params)
data = response.json()

# The historical prices should now be accessible
historical_prices = data['prices']
```

```python
# Extracting timestamp and prices from the historical prices data
timestamps, prices = zip(*historical_prices)

# Creating a DataFrame
btc_df = pd.DataFrame({'Timestamp': timestamps, 'Price': prices})

# Converting timestamp to datetime
btc_df['Timestamp'] = pd.to_datetime(btc_df['Timestamp'], unit='ms')

# Displaying the DataFrame
print(btc_df)
```
